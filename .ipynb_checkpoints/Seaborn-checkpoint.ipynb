### Seaborn

Seaborn est un peu comme un framework de matplotlib

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Les données du titanic
chemin = "./data/titanic.csv"

titanic = pd.read_csv(chemin)
titanic
```

#### Countplot (Analyse univarié variable catégorielle)

```python
#  Avec ça plus besoin forcément de passer par value_counts()
sns.countplot(x='Sex', data=titanic)
plt.title("Sexe des voyageurs du titanic")
plt.show()
```

#### Box plot (Analyse univariée numérique), distribution

```python
#  Box plot (Analyse univariée numérique)
#  univariée : une seule variable qu'on prendet qu'on analyse (autre: bivariée, multivariée)
#  boxplot, la boite à moustache

sns.boxplot(data=titanic, x="Age")
#sns.boxplot(data=titanic, y="Age")
```

```python
titanic_male = titanic[titanic.Sex == 'male']
titanic_fem = titanic[titanic.Sex == 'female']

kwargs = dict(alpha=0.5, bins=20)

plt.hist(titanic_male['Age'], **kwargs, color='blue', label='Male')
plt.hist(titanic_fem['Age'], **kwargs, color='red', label='Female')

plt.gca().set(title='Repartition Homme femme', ylabel='frequence')
plt.legend();
```

```python
sns.kdeplot(titanic["Age"])
```

```python
#  Pour voir la distribution

sns.distplot(titanic["Age"])
```

#### Violin plot(x=category, y=numerique) Analyse bivariée

```python
#  Analyser l'age et la survie

sns.violinplot(x='Survived', y='Age', data=titanic)
```

```python
sns.violinplot(x='Survived', y='Age', data=titanic_male)
```

```python
sns.violinplot(x='Survived', y='Age', data=titanic_fem)
```

#### Analyse multivariée

```python
# Heatmap(plusieurs variables numériques)

titanic.corr()  #  Pour avoir la corrélation entre toutes les variables numériques
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap='YlGnBu')
```

```python
sns.heatmap(titanic.corr()[['Survived', 'Age']], annot=True, cmap='YlGnBu')
```

```python
# FacetGrid(2 variables categorielles et 1 numérique)

cond = (titanic.Sex=='male') & (titanic.Survived==0)
titanic[cond]['Age'].hist()
```

```python
grid = sns.FacetGrid(titanic, col='Sex', row='Survived')
grid.map(plt.hist, 'Age')
```

```python
# Regplot (Reg pour régression linéaire)

sns.scatterplot(data=titanic, x='Age', y='Fare')
```

```python
sns.regplot(data=titanic, x='Age', y='Fare')
```

```python
#  lmplot (facetgrid + reqplot) 2 var catégorielles 2 var numérique
titanic.Fare.hist()
```

```python
sns.lmplot(data=titanic, x='Age', y='Fare', row='Survived', col='Sex')
```

```python
#  Strip plot: scatter mais pour vis une var num et une var cat
sns.stripplot(data=titanic, x='Survived', y='Age')
```

```python
sns.swarmplot(data=titanic, x='Survived', y='Age')
```

```python
#  Catplot = strip plot + facet grid

sns.catplot(data=titanic, x='Survived', y='Age', col='Sex', row='Pclass')
```

```python
sns.catplot(data=titanic, x='Survived', y='Age', col='Sex', row='Pclass', kind='violin')
```

```python
sns.catplot(data=titanic, x='Survived', y='Age', col='Sex', row='Pclass', kind='swarm')
```

```python
#  Pairplot (matrice de visualisation)
# titanic
titanic['Survived'] = titanic['Survived'].astype('category')  # Pour catégoriser Survived au lieu de 0 1
```
