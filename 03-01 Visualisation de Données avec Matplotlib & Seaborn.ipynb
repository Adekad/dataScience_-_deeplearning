```python
%matplotlib inline
```

```python
# Ancienne façon de faire: Matlab
x = [1, 2, 3, 4]
y = [3, 6, 8, 12]
plt.plot(x, y) # Affiche déjà
plt.title('Mon 1er Graphe Matplotlib')
plt.ylim(0, 15)  # Pour avoir un affichage allant jusqu'à 15 sur l'axe des y
plt.xlabel('axe des x')
plt.ylabel('axe des y')
plt.show() # Pour enlever les inscriptions avant et juste avoir le graphe
```

```python
# 2eme façon de faire: Orientée Objet

x = [1, 2, 3, 4]
y = [3, 6, 8, 12]
fig, ax = plt.subplots(figsize=(16, 8), nrows=2, ncols=1) 
# fig.. pour la figure, nrows pour le nombre de ligne et ncols pour le nombre de colone
ax[0].plot(x, y)
ax[0].set_xlabel("1er plot")
ax[1].scatter(np.array(x)+1, y)
ax[1].set_xlabel("2eme plot")
```

```python
# Vu qu'on a 2 maniere de faire généralement on fait une union des 2, en commençant par OO et en poursuivant avec l'autre

fig = plt.figure(figsize=(16, 8))  # Créer une figure

ax1 = fig.add_subplot(211)  # Définir le nombre de subplot et creer les axes
ax2 = fig.add_subplot(212)  # 2 pour 2 rows, 1 pour une col, 2 pour subplot en 2eme position

# Modifier chaque ax pour faire l'affichage
ax1.set_xlabel('axe des x')
ax1.set_ylabel('axe des y')
ax1.plot(x, y)

ax2.set_xlabel('Categorie')
ax2.set_ylabel('Taille')
x2 = ['Homme', 'Femme']
y2 = [180, 160]
ax2.bar(x2, y2)
```

#### Guide de visualisation 

    - une variable numérique : histogramme, boxplot, kernel density plot
    - une variable catégorielle : piechart, barchat
    - 2 variables numériques : scatter plot, heatmap
    - 2 variables catégorielles : barchat
    - 1 variable numérique, 1 variable catégorielle : Violin plot
    - 1 variable numérique, 2 variables catégorielles : Facet grid
