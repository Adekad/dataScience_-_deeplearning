#### Gérer les valeurs manquantes 

```python
data = np.arange(15, 30).reshape(5, 3)
dfx = pd.DataFrame(data, index=['pomme', 'banane', 'citron', 'orange', 'mangue'], 
                  columns=['store1', 'store2', 'store3'])
dfx
```

```python
dfx['store4'] = np.nan
dfx.loc['goyave'] = np.arange(15, 19)
dfx.loc['ananas'] = np.nan
dfx.at['goyave', 'store3'] = np.nan
dfx
```

```python
dfx.isnull()  # isnull et isna retourne même chose(True quand y a des valeur manquante dans une cas),
# notnull pour le contraire
```

```python
dfx.isnull().sum()
```

```python
(dfx.isnull().mean(axis=1)) * 100
```

```python
dfx.dropna()    # Supprimer les valeurs manquantes quand on a au moins une valeur manquante dans la ligne
```

```python
dfx.dropna(how='all') # Supp la ligne ou y a que des val manquante, axis=1 pour col
```

```python
dfx.dropna(thresh=5, axis=1) # Supp la col ou y a au moin 5 val manquantes
```

```python
dfx.fillna(0)   # Remplacer toutes les np.nan par 0
```

```python
dfx.mean()
```

```python
dfx.fillna(dfx.mean())
```

```python
dfx.store4.fillna(method='ffill') # forward filling
```

```python
dfx.store4.fillna(method='bfill') # back filling
```

```python
dfx
```

#### Transformer une variable numérique en variable catégorielle

```python
note = np.random.randint(20, size=(15,))
```

```python
note
```

```python
bins = [-1,10, 12, 15, 18, 20]  # -1 à 10 echoué, 10 à 12 passable ainsi de suite....
```

```python
bins_name = ['échoué', 'passable', 'assez-bien', 'bien', 'très bien']
```

```python
var_cat = pd.cut(note, bins, right=False, labels=bins_name)  # right false pour dire la borne fin est dans l'autre intervalle
```

```python
var_cat
```
