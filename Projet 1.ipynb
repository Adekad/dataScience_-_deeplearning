#### Ville dans laquelle on a le plus vendu de produits

```python
# On a pas de colonne ville mais on a Adresse

df_clean['Purchase Address'].unique()
```

```python
def get_ville(adresse): 
    # Fonction pour  rÃ©cup la ville Ã  partir de l'adresse
    return adresse.split(',')[1].strip()  # split pour une sÃ©paration par , 
# et strip pour enlever les espaces(underscod) de par et d'autre
```

```python
get_ville('760 Church St, San Francisco, CA 94016')
```

```python
# Appliquer la fonction Ã  toute une colonne grÃ¢ce Ã  apply

df_clean['ville'] = df_clean['Purchase Address'].apply(get_ville)
```
